```python
class Doctor:
  def __init__(self,Doctor_ID="", Name="", Specialization="", Working_Time="", Qualification="", Room_Number=""):
    self.Doctor_ID = Doctor_ID
    self.Name = Name
    self.Specialization = Specialization
    self.Working_Time = Working_Time
    self.Qualification = Qualification
    self.Room_Number = Room_Number

  def get_doctor_id(self):
    return self.Doctor_ID
  def get_name(self):
    return self.Name
  def get_specialization(self):
    return self.Specialization
  def get_working_time(self):
    return self.Working_Time
  def get_qualification(self):
    return self.Qualification
  def get_room_number(self):
    return self.Room_Number

  def set_doctor_id(self, New_Doctor_ID):
    self.Doctor_ID = New_Doctor_ID
  def set_name(self, New_Name):
    self.Name = New_Name
  def set_specialization(self, New_Specialization):
    self.Specialization = New_Specialization
  def set_working_time(self, New_Working_Time):
    self.Working_Time = New_Working_Time
  def set_qualification(self, New_Qualification):
    self.Qualification = New_Qualification
  def set_room_number(self, New_Room_Number):
    self.Room_Number = New_Room_Number

  def __str__(self):
    return f"{self.Doctor_ID}_{self.Name}_{self.Specialization}_{self.Working_Time}_{self.Qualification}_{self.Room_Number}"

class DoctorManager:
    def __init__(self):
        self.doctor_list = [] #makes an empty list
        self.read_doctor_file() # reads the file and adds it to the list using the method read doctor file

    def read_doctor_file(self):
        with open('doctors.txt', 'r') as file:
            next(file)  #this is to skip the header in the file
            for line in file:
                line = line.strip()
                if line:  #this checks if the line is not empty
                    parts = line.split("_")
                    if len(parts) == 6:  #this makes sure the line has all 6 parameters
                        doctor = Doctor(parts[0], parts[1], parts[2], parts[3], parts[4], parts[5]) # all the parameters are assigned to doctor here
                        self.doctor_list.append(doctor)  #this will append the object doctor to the list

    def format_dr_info(self,doctor): #will get the properties using the getters methods
        doctor_id = doctor.get_doctor_id()
        name = doctor.get_name()
        specialization = doctor.get_specialization()
        working_time = doctor.get_working_time()
        qualification = doctor.get_qualification()
        room_number = doctor.get_room_number()
        formatted_info = f"{doctor_id}_{name}_{specialization}_{working_time}_{qualification}_{room_number}" #format the info with _ in between
        return formatted_info

    def enter_dr_info(self):    #takes the info from user and return it as a doctor object
        Doctor_ID = input('Enter Doctor ID: ')
        Name = input('Enter Doctor Name: ')
        Specialization = input('Enter Doctor Specialization: ')
        Working_Time = input('Enter Doctor Working Time: ')
        Qualification = input('Enter Doctor Qualification: ')
        Room_Number = input('Enter Doctor Room Number: ')
        return Doctor(Doctor_ID, Name, Specialization, Working_Time, Qualification, Room_Number)

    def search_doctor_by_id(self):   # asks for id then does a for loop to look for that ID in the list, when found it will can the display_dr_info method
        search_ID= input("Enter the Doctor's ID you wish to search for")
        found = False
        for doctor in self.doctor_list:
            if doctor.get_doctor_id() == search_ID:
                print("\nId   Name                   Speciality      Timing          Qualification   Room Number\n")  #header
                print(self.display_dr_info(doctor))
                found = True
                break
        if not found:
            print(f"Doctor with ID {search_ID} not found.")


    def search_doctor_by_name(self): #same as the ID one but with the name
        search_name = input("Enter the Doctor's Name you wish to search for")
        found = False
        for doctor in self.doctor_list:
            if doctor.get_name() == search_name:
                print("\nId   Name                   Speciality      Timing          Qualification   Room Number\n")  #header
                print(self.display_dr_info(doctor))
                found = True
                break
        if not found:
            print(f"Doctor with name {search_name} not found.")

    def display_dr_info(self, doctor):# displays the info of a doctor in a formatted way
        return f'''{doctor.get_doctor_id():<5} {doctor.get_name():<23} {doctor.get_specialization():<15} {doctor.get_working_time():<15} {doctor.get_qualification():<15} {doctor.get_room_number():<12}'''

    def edit_dr_info(self): #this uses the setters from the Doctor class to change the info of a doctor in the list and then writes it to the file
        search_ID = input("Enter the Doctor's ID you wish to edit")
        found = False
        for doctor in self.doctor_list:
            if doctor.get_doctor_id() == search_ID:
                doctor.set_name(input("enter new name: "))
                doctor.set_specialization(input("enter new specialization: "))
                doctor.set_working_time(input("enter new working time: "))
                doctor.set_qualification(input("enter new qualification: "))
                doctor.set_room_number(input("enter new room number: "))
                self.write_list_of_doctors_to_file()
                print (f"Doctor with ID {search_ID} has been updated")
                found = True
                break
        if not found:
            print(f"Doctor with ID {search_ID} not found.")

    def display_doctors_list(self):  #this first prints a header then will print the doctor info using the method display_dr_info one by one .
        print("Id   Name                   Speciality        Timing        Qualification   Room Number\n")
        for doctor in self.doctor_list:
            print(self.display_dr_info(doctor))


    def write_list_of_doctors_to_file(self): #this will open the file, then writes the header there , after that it will run a for loop on the dr list and user the format_dr_info mehtod to write the info to the file.
        with open('doctors.txt', 'w') as file:
            file.write("id_name_specilist_timing_qualification_roomNb\n")
            for doctor in self.doctor_list:
                formatted_info = self.format_dr_info(doctor)
                file.write(formatted_info + '\n')
            with open('doctors.txt', 'a') as file:
                file.write('\n')  #this makes sure the file is starting in a new line as otherwise it causes issues in the code.
        print("Doctor list saved successfully.")

    def add_dr_to_file(self):
        new_doctor = self.enter_dr_info()
        self.doctor_list.append(new_doctor)
        formatted_info = self.format_dr_info(new_doctor)
        with open('doctors.txt', 'a') as file:
            file.write(formatted_info + '\n')  #this makes sure the file is starting in a new line as otherwise it causes issues in the code
        print("Doctor added successfully.")


```
